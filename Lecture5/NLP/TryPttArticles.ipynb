```python
import re

def get_valid_user_id(uid):
    re_chech = re.search("([a-zA-Z0-9]+)",uid)
    if re_chech == None:
        return False
    else:
        return True

def get_invalid_user_id(uid):
    re_chech = re.search("([a-zA-Z0-9]+)",uid)
    if re_chech == None:
        return True
    else:
        return False
    

article_df[article_df["user_id"].map(get_invalid_user_id)]
```

```python
from pymongo import MongoClient
mongo_client = MongoClient()
ptt_crawler_db = mongo_client.ptt_crawler_db
ptt_article = ptt_crawler_db.ptt_article
ptt_article_push = ptt_crawler_db.ptt_article_push
```

```python
import pandas as pd 

articles_df = pd.DataFrame(list(ptt_article.find({},{"url":1,"user_id":1,"time":1})))
articles_df["board"] = articles_df["url"].map(lambda xx:xx.split("http://www.ptt.cc/bbs/")[1].split("/")[0])
```

```python
import datetime
import re
import numpy as np

def convert_timestr(xx):
    try:
        
        
        yy = re.search(ur"(:?Mon|Tue|Wed|Thu|Fri|Sat|Sun)\s*(:?Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s*([0-9]+)\s*([0-9+]+)\s*:\s*([0-9]{2})\s*:\s*([0-9]{2})\s*([0-9]{4})",
                       xx)
            
        if yy != None:
            temp = datetime.datetime.strptime(" ".join(yy.groups()),"%a %b %d %H %M %S %Y")
             
        else:
            yy = re.search(ur"(?P<m>[0-1]*[0-9])/(?P<d>[0-9]+)/(?P<y>[0-9]{4})\s*(?P<H>[0-9+]+)\s*:\s*(?P<M>[0-9]+)\s*:\s*(?P<S>[0-9]+)","01/26/2005 12:34:22 Wed").groupdict()
        
            temp = datetime.datetime(int(yy["y"]),
                                     int(yy["m"]),
                                     int(yy["d"]),
                                     int(yy["H"]),
                                     int(yy["M"]),
                                     int(yy["S"]))
#        print "[successfully] yy = ",yy
        return temp
    
    except Exception as e:
#    except:
        if np.isnan(xx):
            pass
#            print "xx == np.nan"
        else:
            print "Error = ",e
            print "[fail] xx = ",xx
            print  "Error Formate:%s" % xx
        return "Error Formate"


print articles_df["time"][1:1000].map(convert_timestr)
```

```python
test_convert_time = articles_df["time"].map(convert_timestr)
test_convert_time.head()
```

```python
articles_filtered_df = articles_df[test_convert_time  != "Error Formate"]
articles_filtered_df["datetime"] = test_convert_time[test_convert_time  != "Error Formate"]
```

```python
articles_filtered_df.head()
```

```python
articles_filtered_df.set_index("datetime")

articles_filtered_df.index = articles_filtered_df["datetime"]

articles_clean_df = articles_filtered_df[["user_id","board","_id"]]

articles_clean_df.groupby([lambda x: x.year, lambda x: x.month,"board","user_id"]).size()

print articles_clean_df["board"].unique()
```

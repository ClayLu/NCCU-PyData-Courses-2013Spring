```python
import numpy as np
from PIL import Image
import matplotlib.cm as cm
im = Image.open("coffee.jpg").convert('L')
im_array = pylab.array(im)
im_array = im_array[:-1,:]
print im_array.shape
im_array_n = im_array + 200*np.random.randn(*im_array.shape)

im_array_t = np.fft.fftshift(np.fft.fft2(im_array))
im_array_n_t = np.fft.fftshift(np.fft.fft2(im_array_n))
kx = np.arange(1,im_array_n_t.shape[1]+1,1)
ky = np.arange(1,im_array_n_t.shape[0]+1,1)
KX,KY = np.meshgrid(kx,ky)
GaussFilter = np.exp(-0.0001*(KX - im_array_n_t.shape[1] / 2)**2 - 0.0001*(KY - im_array_n_t.shape[0] / 2)**2)

im_array_n_t_F = np.multiply(im_array_n_t,GaussFilter)
im_array_n_t_F_ifft = np.fft.ifft2(np.fft.ifftshift(im_array_n_t_F)).real
im_array_t_ifft = np.fft.ifft2(np.fft.ifftshift(im_array_t)).real
im_array_n_t_ifft = np.fft.ifft2(np.fft.ifftshift(im_array_n_t)).real
```

```python
# pylab.imshow(im_array_n_t_F_ifft,cmap="hot")
pylab.imshow(im_array_n_t_F_ifft,cmap=cm.gray)
```

-   ## Change the point of View (Using FFT)
-   ## Separating Noise from Information (Using Gaussian Filter)
-   ## Change back to the original point of View (Using iFFT)

# Linear Programming:

## $$ \begin{eqnarray} \min_{x_1,x_2} x_1 & + & x_2 \\ 2x_1 & + & x_2 & \geq 1 \\ x_1 & + & 3x_2 & \geq 1 \\ x_1 &&& \geq 0 \\ x_2 &&& \geq 0 \\ \end{eqnarray}$$

```python
xx1 = np.arange(0, 1, 0.001)
xx2 = np.arange(0, 1, 0.001)
XX1, XX2 = np.meshgrid(xx1, xx2)

obj_fn =  XX1 + XX2
c1_fn = 2 * XX1 +  XX2
c2_fn = XX1 + 3 * XX2

c_region = obj_fn.copy()

c_region[(c1_fn > 1) & (c2_fn > 1)] = 200
c_region[(c1_fn < 1) | (c2_fn < 1)] = 255


fig = pylab.figure()

c2_fn_plt = pylab.imshow(c_region,origin=[0,0],cmap=cm.Greys_r)
pylab.xticks(np.linspace(0,len(xx1),11),np.linspace(0,1,11))
pylab.yticks(np.linspace(0,len(xx2),11),np.linspace(0,1,11))


obj_fn_values = np.arange(0,1,0.1)
obj_fn_plt = pylab.contour(obj_fn,obj_fn_values,colors='g')
pylab.clabel(obj_fn_plt, inline=1, fontsize=10)

ax, = fig.get_axes()
ax.text(0.8*len(xx1), 0.05*len(xx1), r"$x_1 + 3x_2 = 1$", fontsize=20, color="b")
ax.text(0.05*len(xx1), 0.9*len(xx1), r"$2x_1 + x_2 = 1$", fontsize=20, color="b")
ax.text(0.8*len(xx1), 0.9*len(xx1), r"$2x_1 + x_2 \geq 1$", fontsize=20)
ax.text(0.8*len(xx1), 0.8*len(xx1), r"$x_1 + 3x_2 \geq 1$", fontsize=20)

fig.savefig("LP_Demo.png",figsize=(20,15),dpi=200)
```

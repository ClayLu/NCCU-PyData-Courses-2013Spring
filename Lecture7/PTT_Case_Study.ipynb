```python
import datetime
import re
import numpy as np

def convert_timestr(xx):
    try:
        
        
        yy = re.search(ur"(:?Mon|Tue|Wed|Thu|Fri|Sat|Sun)\s*(:?Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s*([0-9]+)\s*([0-9+]+)\s*:\s*([0-9]{2})\s*:\s*([0-9]{2})\s*([0-9]{4})",
                       xx)
            
        if yy != None:
            temp = datetime.datetime.strptime(" ".join(yy.groups()),"%a %b %d %H %M %S %Y")
             
        else:
            yy = re.search(ur"(?P<m>[0-1]*[0-9])/(?P<d>[0-9]+)/(?P<y>[0-9]{4})\s*(?P<H>[0-9+]+)\s*:\s*(?P<M>[0-9]+)\s*:\s*(?P<S>[0-9]+)","01/26/2005 12:34:22 Wed").groupdict()
        
            temp = datetime.datetime(int(yy["y"]),
                                     int(yy["m"]),
                                     int(yy["d"]),
                                     int(yy["H"]),
                                     int(yy["M"]),
                                     int(yy["S"]))
#        print "[successfully] yy = ",yy
        return temp
    
    except Exception as e:
#    except:
        if np.isnan(xx):
            pass
#            print "xx == np.nan"
        else:
            print "Error = ",e
            print "[fail] xx = ",xx
            print  "Error Formate:%s" % xx
        return "Error Formate"


print articles_df["time"][1:1000].map(convert_timestr)
```

```python
post_viewed_by_user_df = merge_push_info_df.groupby(["url"]).apply(join_user).reset_index()
post_viewed_by_user_df.columns = ["post","users"]
post_viewed_by_user_df.head()
```

```python
import PlaYnlp.tokenizer as tkr
import PlaYnlp.vectorizer as vcr
```

```python
def view_tokenize(text):
    if isinstance(text,(unicode,str)):
        return text.split("//")
    else:
        return []
    
vect_gen_init_kwargs = {"tokenizer":view_tokenize,"lowercase":False}
```

```python

test_sdtm = vcr.vectorize_text(df=post_viewed_by_user_df,
                                text_col="users", 
                                idx_col="post", 
                                cond_query={},
                                idx_query= [],
                                vect_gen=vcr.CountVectorizer, 
                                vect_gen_init_kwargs = vect_gen_init_kwargs)
```

```python
L0_norm_summarizer=lambda xx:xx.sign().sum(axis=0)
```

```python
post_idx = 2000
print test_sdtm['row_idx'][post_idx]
pushed_user_idx = np.nonzero(test_sdtm["sdtm"][post_idx,:])[1]
print pushed_user_idx

related_posts_idx = (test_sdtm.sub_sdtm(select_col_idx = pushed_user_idx).T.summarize_sdtm(L0_norm_summarizer)>0)['summary_data']
print "related_posts_idx = ",related_posts_idx

top_posts_idx = test_sdtm.sub_sdtm(select_row_idx = related_posts_idx).T.summarize_sdtm(L0_norm_summarizer)['summary_data'].argsort()[-10:]
print "top_posts_idx = ",top_posts_idx


related_posts_sdtm_summary = test_sdtm.sub_sdtm(select_row_idx = related_posts_idx).T.summarize_sdtm(L0_norm_summarizer)
related_posts_sdtm_summary['terms_idx'][top_posts_idx]
```

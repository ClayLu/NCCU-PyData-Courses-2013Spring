## 把上述全部動作寫成一個 function

給定 yahoo Stock id，回傳整理好的 `df`

```python
def get_stock_major_info(stock_id):
    """return info of given stock id"""
    # get original df
    res = requests.get("http://tw.stock.yahoo.com/d/s/major_%s.html" % stock_id)
    tables = pd.read_html(res.text)
    
    want_data_table = [tab for tab in tables if tab.shape == (16, 8)][0]
    want_data_table = want_data_table.iloc[1:, :]
    tab_left_half = want_data_table.iloc[:, :4].values
    tab_right_half = want_data_table.iloc[:, 4:].values
    df = pd.DataFrame(
        np.r_[tab_left_half, tab_right_half]
    )
    
    # give it date, stock info
    df.columns = ['Name', 'Buy', 'Sell', 'Net']
    date_str = tables[0].iat[0, 0]
    re_m = re.match(
        u"資料日期：(\d+)[ /]*(\d+)[ /]*(\d+)",
        date_str, re.UNICODE
    )
    date_splitted = map(
        int,  # 一個可以把字串轉整數的函式
        re_m.groups()   #被切割好的 民國年, 月, 日
    )
    date_splitted[0] += 1911  # 把民國轉為西元
    dt_obj = datetime.date(*date_splitted)

    df['Date'] = dt_obj  # 塞 datetime 的物件
    df['Stock_id'] = stock_id
    
    # reorder
    col_names = df.columns.values
    df = df[np.r_[col_names[-2:], col_names[:-2]]]
    
    return df
```
